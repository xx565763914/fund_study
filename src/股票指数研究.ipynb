```python
def get_data_from_yahoo(code: str, start: datetime.datetime, end: datetime.datetime):
    current_time = start
    
    frame_list = []
    while current_time < end:
        
        print(current_time, min(end, current_time + datetime.timedelta(days=365)))
        frame = web.DataReader(code, "yahoo", current_time, min(end, current_time + datetime.timedelta(days=365)))
        frame_list.append(frame)
        current_time = min(end, current_time + datetime.timedelta(days=365))
        print(current_time)
        time.sleep(0.5)
    return pd.concat(frame_list)
```

```python
result = get_data_from_yahoo('000300.ss', start, end)
```
